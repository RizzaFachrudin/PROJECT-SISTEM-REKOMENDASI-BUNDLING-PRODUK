## Menghitung Frekuensi Produk

Menghitung jumlah kemunculan setiap produk.

```python
from collections import Counter

all_items = []

for item in df['Items']:
    products = item.split(',')

    for p in products:
        all_items.append(
            p.strip().upper()
        )

counter = Counter(all_items)

counter.most_common(10)
```

### Interpretasi

Output berikut menunjukkan 10 produk yang paling sering muncul dalam transaksi.

## Visualisasi Produk Terlaris

```python
import matplotlib.pyplot as plt

top_products = counter.most_common(10)

produk = [x[0] for x in top_products]
jumlah = [x[1] for x in top_products]

plt.figure(figsize=(10,5))

plt.bar(produk, jumlah)

plt.title("10 Produk Terlaris")

plt.xticks(rotation=45)

plt.show()
```

### Interpretasi Hasil

Grafik menunjukkan produk yang paling sering dibeli pelanggan dan dapat dijadikan fokus promosi.
